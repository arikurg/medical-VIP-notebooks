## Section 1: Introduction to NumPy
NumPy is a library for numerical operations in Python, offering fast and efficient array computations.

```python
import numpy as np

# Create a NumPy array
arr = np.array([1, 2, 3, 4, 5])
print('Array:', arr)
print('Sum:', arr.sum())
print('Mean:', arr.mean())

# basic ops
print("add 2:", arr + 2)
print("elementwise *2:", arr * 2)
```

We can have multi-dimensional arrays

```python
# Shapes
# matrix (2D) and shape
M = np.array([[1, 2, 3],
              [4, 5, 6]])
print("M:", M)
print("M shape:", M.shape)
print("M sum:", M.sum())
print("M sum axis=0 (columns):", M.sum(axis=0))
print("M mean axis=1 (rows):", M.mean(axis=1))
```

Manipulating and Masking Arrays


```python
# Masking
mask = M > 3
print("mask:\n", mask)
print("elements > 3:", M[mask])

# np.where example 
indices = np.where(M % 2 == 0)                      # finds indicies in M where elements are even
print("indices of even element:", indices)          # tuple of (rows, cols) that match the criteria 

# np.where with conditional example
results = np.where(M <= 3, M * 2, M)                # if element is <= 3, double it, else, do nothing to element
print("Modified M matrix: \n", results)
```

## Section 3: Introduction to Pandas
Pandas is a library for data manipulation and analysis. It uses `DataFrames` and `Series` as its main structures. Think of Pandas Dataframes as massive excel tables. The difference is that now we interact with them programattically.

```python
import pandas as pd

# Create a DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35], 'City': ['NY', 'LA', 'SF']}
df = pd.DataFrame(data)
print(df)
```

## Section 4: Basic Plotting
Visualization helps you understand your data. We'll use `matplotlib` for basic plots.

```python
import matplotlib.pyplot as plt

# Line plot
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 17, 20]
plt.plot(x, y, marker='o')
plt.title('Line Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

**Putting it all together: plotting a bar chart from our sample.csv data!**

```python
counts = df['Cholesterol Level'].value_counts()
bars = plt.bar(counts.index, counts.values, color=['skyblue', 'lightgreen', 'salmon']) # sets the index to place the bar, the value, and color
plt.xlabel('Cholesterol Level')                                                        # X axis label
plt.ylabel('Count')                                                                    # Y axis label
plt.title('Distribution of Blood Pressure Categories')                                 # Title
plt.bar_label(bars)                                                                    # Places count above the bar
# plt.savefig("Blood Pressure Chart.png")                                              # Saving the plot as an png file
plt.show()                                                                             # Displaying a plot
```
